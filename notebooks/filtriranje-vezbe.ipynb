## Moving average

Moving average je tehnika koja se koristi za "uglačavanje" (smoothing) signala. Ideja je da imamo neku fiksiranu širinu "prozora" $w$ i da idemo redom kroz niz (signal). Svaku vrednost signala zamenjujemo prosečnom vrednošću signala u narednih w tačaka.

### Primer

Imamo signal $[1,2,3,4,5]$ i širinu prozora $2$. Prvi element (1) ćemo zameniti sa prosekom $w = 2$ narednih vrednosti signala (počev od trenutne), odnosno sa $(1 + 2) / 2 = 1.5$.

Taj proces ponavaljamo za sve vrednosti signala, dok je moguće odabrati narednih $w$ vrednosti. 
Rezultujući signal u ovom slučaju će biti $[1.5, 2.5, 3.5, 4.5]$

Kao što se može videti, dužina signala se smanjila za 1. U opštem slučaju, rezultujući signal je kraći za $w-1$. 

### Zadatak 

Implementirati funkciju movingAverage koja realizuje objašnjenu tehniku uglačavanja

```python
def movingAverage(x: np.array, w: int) -> np.array:
    smoothed = np.zeros(len(x) - w + 1)
    # ovde je potrebno implementirati zadatu funkciju
    
    return smoothed

Testbench(movingAverage)

# ucitavanje jednog od signala
fs, x = wavfile.read('datasets/sounds/01_guitar_E2.wav')
N = len(x)

# plotovanje originalnog signala

n = np.arange(N//10, N//10 + 2000)
fig = plt.figure(figsize = [6, 4])
n = np.arange(N//10, N//10 + 2000)
t = n/fs
plt.plot(t, x[n])
plt.subplots_adjust(bottom=0.15, left=0.15)
plt.xlabel(r'$t$ [s]')
plt.ylabel(r'$x(t)$')


# plotovanje uglačanog signala

smoothed = movingAverage(x, 100)
N1 = len(smoothed)
fig = plt.figure(figsize = [6, 4])
n1 = np.arange(N1//10, N1//10 + 2000)
t1 = n1/fs
plt.plot(t1, smoothed[n1])
plt.subplots_adjust(bottom=0.15, left=0.15)
plt.xlabel(r'$t$ [s]')
plt.ylabel(r'$x(t)$')
```
